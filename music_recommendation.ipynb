# data processing...

```python
data=data.dropna(axis=0)
```

```python
data= pd.concat([data,pd.get_dummies(data['year'], prefix='year')],axis=1)
```

```python
data= pd.concat([data,pd.get_dummies(data['artist_name'], prefix='artist_name')],axis=1)
```

```python
data= pd.concat([data,pd.get_dummies(data['release'], prefix='release')],axis=1)
```

```python
data.drop(columns=['artist_name','release','year'],inplace=True)
```

```python
data
```

```python
data1=data.copy()
```

```python
data1.drop(columns=['song_id','title'],inplace=True)
```

```python
data1
```

```python
data1.info()
```

# cosine_similarity.....

```python
values=[]
index=[]
i=1
for j in range(1,10000):
    x=data1.iloc[i]
    y=data1.iloc[j]
    z=sum(x*y)
    w=np.sqrt(sum(x*x))+np.sqrt(sum(y*y))
    k=(z/w)
    values.append(k)
    index.append(j)
```

# evaluation.

```python
best_match=[]
for i in values1:
    if i>0:
        best_match.append(i)
```

```python
j=[]
for i in best_match:
    j.append(values.index(i))
print(set(j))
    
```
