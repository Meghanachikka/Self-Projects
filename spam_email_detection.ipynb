```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
pd_data['text']=pd_data['text'].apply(word_tokenize)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
values=[]
for i in pd_data['text']:
    lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in i])
    values.append(lemmatized_output)
```

```python
pd_data['text']=values
```

```python
pd_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer='word')
vectorizer.fit(pd_data["text"])
```

```python
print(vectorizer.vocabulary_)
```

```python
vector = vectorizer.fit_transform(pd_data['text'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(vector,pd_data['text'],test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_predict=clf.predict(x_test)
print('accuracy',round(accuracy_score(y_test,y_predict)*100,4),'%')
print('confusion matrix\n',confusion_matrix(y_test,y_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 10)
clf.fit(x_train,y_train)
predict2 = clf.predict(x_test)
print(accuracy_score(y_test,predict2))
```
