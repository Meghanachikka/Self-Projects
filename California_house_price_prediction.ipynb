#### Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

#### Importing Dataset

```python
data=pd.read_csv('cf.csv')
```

```python
data.head()
```

#### Data analytics

```python
data.info()
```

```python
data.describe()
```

#### Training (using mean of median_house_value)

```python
data['status']=data['median_house_value']<207300
data.head()
```

```python
data['status'].replace([True,False],['affordable','not affordable'],inplace=True)
data.head()
```

#### Plotting histogram of Status

```python
data['status'].hist()
```

#### Ploting 'Status' using CatPlot

```python
sns.catplot(x='median_house_value',y='status',kind='box',data=data)
plt.show()
```

#### Ploting 'Status' using HeatMap

```python
data.corr()
```

```python
sns.heatmap(data.corr(),square=True,vmax=0.8,vmin=-0.8)
plt.show()
```

## Conclusion

#### If  'median_house_value'  is less than 207300, the house is 'affordable'

#### If  'median_house_value'  is greater than 207300, the house is 'not affordable'
