# data preprocessing....

```python
train['class'].replace(['normal','anomaly'],[0,1],inplace=True)
```

```python
train=train.drop(columns=['duration','dst_bytes','land','wrong_fragment','urgent','hot','num_outbound_cmds','is_host_login','dst_host_count','dst_host_diff_srv_rate','dst_host_serror_rate','dst_host_rerror_rate'])
```

```python
from sklearn.model_selection import train_test_split
train_x,val_x,train_y,val_y = train_test_split(train.drop(columns=['class','protocol_type','service','flag']),train['class'],test_size=0.1)
```

# training


decision tree classifier...

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(train_x,train_y)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
y_predict=clf.predict(val_x)
print('accuracy',round(accuracy_score(val_y,y_predict)*100,4),'%')
print('confusion matrix\n',confusion_matrix(val_y,y_predict))
print(classification_report(val_y,y_predict))
```

random forest classifier..

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=10)
clf.fit(train_x,train_y)
predict2=clf.predict(val_x)
print('accuracy',round(accuracy_score(val_y,predict2)*100,4),'%')
print(classification_report(val_y,predict2))
```

kneighborsclassifier....

```python
from sklearn.neighbors import KNeighborsClassifier
clf2 = KNeighborsClassifier()
clf2.fit(train_x, train_y)
predict4=clf2.predict(val_x)
print('accuracy',round(accuracy_score(val_y,predict4)*100,4),'%')
print(classification_report(val_y,predict4))
```

linearregression...

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(train_x,train_y)
predict5=reg.predict(train_x)
reg.score(train_x,train_y)
```

# evaluation...

```python
print('accuracy',round(accuracy_score(val_y,y_predict)*100,4),'%..decision tree classifier...')
print('accuracy',round(accuracy_score(val_y,predict2)*100,4),'%..random forest classifier..')
print('accuracy',round(accuracy_score(val_y,predict4)*100,4),'%..kneighborsclassifier....')
print(reg.score(train_x,train_y),'%..linear regresssion....')
```
