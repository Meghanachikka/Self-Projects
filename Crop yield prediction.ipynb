```python
data= pd.concat([data,pd.get_dummies(data['State_Name'], prefix='State_Name')],axis=1)
```

```python
data=pd.concat([data,pd.get_dummies(data['District_Name'],prefix='District_Name')],axis=1)
data=pd.concat([data,pd.get_dummies(data['Season'],prefix='Season')],axis=1)
data=pd.concat([data,pd.get_dummies(data['Crop'],prefix='Crop')],axis=1)
```

```python
data.drop(columns=['State_Name','District_Name','Season','Crop'],inplace=True)
```

```python
data = data.dropna(axis = 0)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data.drop(columns=['Production']),data['Production'],test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
clf=LinearRegression()
clf.fit(x_train,y_train)
```

```python
predict=clf.predict(x_test)
clf.score(x_train,y_train)
```

```python
reg1= Ridge(alpha=2.2)
reg2= Lasso(alpha=2.0)

reg1.fit(x_train,y_train)
reg2.fit(x_train,y_train)

```

```python
predict1=reg1.predict(x_test)
reg1.score(x_train,y_train)
```

```python
predict2=reg2.predict(x_test)
reg2.score(x_train,y_train)
```
